```python
# Apply Price Sanity Checks
aapl_stock["high_price"] = aapl_stock[
    ["high_price", "open_price", "close_price"]
].max(axis=1)

aapl_stock["low_price"] = aapl_stock[
    ["low_price", "open_price", "close_price"]
].min(axis=1)
```

```python
# Apply Price Sanity Checks
goog_stock["high_price"] = goog_stock[
    ["high_price", "open_price", "close_price"]
].max(axis=1)

goog_stock["low_price"] = goog_stock[
    ["low_price", "open_price", "close_price"]
].min(axis=1)
```

```python
# Apply Price Sanity Checks
msft_stock["high_price"] = msft_stock[
    ["high_price", "open_price", "close_price"]
].max(axis=1)

msft_stock["low_price"] = msft_stock[
    ["low_price", "open_price", "close_price"]
].min(axis=1)
```

```python
#Calculate daily return
aapl_stock["daily_return"] = (
    (aapl_stock["close_price"] - aapl_stock["open_price"])
    / aapl_stock["open_price"]
)
aapl_stock
```

```python
#Calculate daily return
goog_stock["daily_return"] = (
    (goog_stock["close_price"] - goog_stock["open_price"])
    / goog_stock["open_price"]
)
goog_stock
```

```python
#Calculate daily return
msft_stock["daily_return"] = (
    (msft_stock["close_price"] - msft_stock["open_price"])
    / msft_stock["open_price"]
)
msft_stock
```

```python
def up_down_summary(df):
    up_days = (df['close_price'] > df['open_price']).sum()
    down_days = (df['close_price'] < df['open_price']).sum()
    ratio = up_days / down_days if down_days != 0 else 0

    return up_days, down_days, ratio
```

```python
results = []

for stock_df in [aapl_stock, msft_stock, goog_stock]:
    company = stock_df['symbol'].iloc[0]   # AAPL / MSFT / GOOG
    
    up, down, ratio = up_down_summary(stock_df)

    results.append({
        "company": company,
        "up_days": up,
        "down_days": down,
        "up_down_ratio": ratio
    })

trend_df = pd.DataFrame(results)
print(trend_df)
```

```python
x = range(len(trend_df))

plt.figure(figsize=(9, 5))
plt.bar(x, trend_df['up_days'], width=0.4, label='UP')
plt.bar([i + 0.4 for i in x], trend_df['down_days'], width=0.4, label='DOWN')

plt.xticks([i + 0.2 for i in x], trend_df['company'])
plt.xlabel("Company")
plt.ylabel("Number of Days")
plt.title("UP vs DOWN Trend per Company")
plt.legend()
plt.tight_layout()
plt.show()
```
